```python
# Standard library imports
import os
import glob
import shutil
import logging
import warnings
from typing import Optional, Union, Dict, List
from datetime import datetime

# Third-party library imports
import pandas as pd
from dateparser import parse
```

```python
# map transactions to core data model
amex_map = {
    'Date Processed': 'date',
    'Description': 'description',
    'Amount': 'amount'
}

nab_map = {
    'Date': 'date',
    'Transaction Details': 'description',
    'Amount': 'amount'
}
```

```python
def stg_tx_data(raw_tx_data_path, cdm_map, provider):
    raw_tx = pd.read_csv(raw_tx_data_path)
    stg_tx = raw_tx[cdm_map.keys()].rename(columns=cdm_map)
    stg_tx['provider'] = provider
    return stg_tx
```

```python
tx_dir = 'C:/Users/Barry/OneDrive/Documents/My stuff/Personal finance/expense_manager/data/tx_data/'
amex_path = os.path.join(tx_dir, 'input/AMEX Transactions 20240120-20240526.csv')
nab_path =os.path.join(tx_dir, 'input/NAB Transactions 20230701-20240526.csv')

stg_amex = stg_tx_data(amex_path, amex_map, provider='AMEX')
stg_nab = stg_tx_data(nab_path, nab_map, provider='NAB')

stg_tx = pd.concat([stg_amex, stg_nab])
```
